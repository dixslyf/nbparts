```python
import matplotlib.pyplot as plt
import math
```

```python
xs = list(x * (2 * math.pi / 100) for x in range(0, 100))
ys = [math.sin(x) for x in xs]

print("This is a plot:")
plt.figure()
plt.plot(xs, ys)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
